### 1만개

```python
import pandas as pd
import re
from kospellpy import spell_init
import requests
import time

spell_checker = spell_init()

def contains_korean(text):
    return bool(re.search("[가-힣]", text))

def reliable_spell_checker(text, retries=3, delay=5):
    for i in range(retries):
        try:
            return spell_checker(text)
        except requests.exceptions.Timeout:
            print(f"Timeout occurred. Retry {i+1}/{retries} after {delay} seconds...")
            time.sleep(delay)
        except requests.exceptions.RequestException as e:
            print(f"Request failed: {e}")
            return text
    print(f"Failed after {retries} attempts. Returning original text.")
    return text

# 첫 번째 1만 개 처리
review_contents = skincare_all_df_original['review_content'][:10000]
corrected_contents = []

for content in review_contents[:10000]:
    try:
        if contains_korean(content):
            corrected_text = reliable_spell_checker(content)
        else:
            corrected_text = content
        corrected_contents.append(corrected_text)
    except ValueError as e:
        print(f"Error processing content: {content[:30]}...") 
        print(f"Error details: {e}")
        corrected_contents.append(content)

skincare_all_df_original.loc[:9999, 'corrected_review_content'] = corrected_contents
skincare_all_df_original.iloc[:10000].to_csv('skincare_corrected_review_content_df_1.csv', index=False)
```

### 3만개

```python
import pandas as pd
import re
from kospellpy import spell_init
import requests
import time

spell_checker = spell_init()

def contains_korean(text):
    if isinstance(text, str):
        return bool(re.search("[가-힣]", text))
    return False #문자열이 아닌 경우 한글이 없다고 판단


def reliable_spell_checker(text, retries=3, delay=5):
    for i in range(retries):
        try:
            return spell_checker(text)
        except requests.exceptions.Timeout:
            print(f"Timeout occurred. Retry {i+1}/{retries} after {delay} seconds...")
            time.sleep(delay)
        except requests.exceptions.ConnectionError as ce:
            print(f"Connection error occurred: {ce}. Retry {i+1}/{retries} after {delay} seconds...")
            time.sleep(delay)
        except requests.exceptions.RequestException as e:
            print(f"Request failed: {e}")
            return text
    print(f"Failed after {retries} attempts. Returning original text.")
    return text

# 세 번째 1만 개 처리 (20000부터 끝까지)
review_contents = skincare_all_df_original['review_content'][20000:]  # 끝 인덱스를 지정하지 않음
corrected_contents = []

for content in review_contents:
    try:
        if contains_korean(content):
            corrected_text = reliable_spell_checker(content)
        else:
            corrected_text = content
        corrected_contents.append(corrected_text)
    except ValueError as e:
        print(f"Error processing content: {content[:30]}...") 
        print(f"Error details: {e}")
        corrected_contents.append(content)

skincare_all_df_original.loc[20000:20000+len(corrected_contents)-1, 'corrected_review_content'] = corrected_contents
skincare_all_df_original.iloc[20000:].to_csv('skincare_corrected_review_content_df_3만_끝.csv', index=False)
```
