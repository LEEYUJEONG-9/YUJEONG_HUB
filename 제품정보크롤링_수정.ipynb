# 제품 정보 크롤링 및 파일 저장

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import numpy as np
import time
import warnings
np.set_printoptions(threshold=np.inf, linewidth=np.inf)
warnings.filterwarnings('ignore')

# 크롬 드라이버 초기화
driver = webdriver.Chrome()

# URL 5개 파일 합치기
dfs = [skincare_id_url, maskpack_id_url, cleansing_id_url, suncare_id_url, bodycare_id_url]
all_url = pd.concat(dfs, ignore_index=True)

# 각 데이터 저장 리스트 초기화
product_id_list = []
category_list1 = []
category_list2 = []
category_list3 = []
prd_name_list = []
price1_list = []
price2_list = []
evg_review_list = []
cnt_review_list = []
review_score_5_list = []
review_score_4_list = []
review_score_3_list = []
review_score_2_list = []
review_score_1_list = []
poll_type_list1_1 = []
poll_type_list1_2 = []
poll_type_list1_3 = []
poll_type_list2_1 = []
poll_type_list2_2 = []
poll_type_list2_3 = []
poll_type_list3_1 = []
poll_type_list3_2 = []
poll_type_list3_3 = []

# 제품 정보를 수집하는 함수
def get_product_info(url):
    try:
        # 올리브영 URL 열기
        driver.get(url)
        


        # 카테고리 가져오기
        category1 = driver.find_element(By.CSS_SELECTOR, '#midCatNm').text
        category2 = driver.find_element(By.CSS_SELECTOR, '#smlCatNm').text
        category3 = driver.find_element(By.CSS_SELECTOR, '#dtlCatNm').text
        category_list1.append(category1)
        category_list2.append(category2)
        category_list3.append(category3)
        time.sleep(5)

        # 제품명 가져오기
        prd_name = driver.find_element(By.CLASS_NAME, 'prd_name').text
        prd_name_list.append(prd_name)

        # 할인 전 가격 가져오기
        try:
            price1 = driver.find_element(By.CLASS_NAME, 'price-1').text.replace('\n원', "")
            price1_list.append(price1)
        except: #할인없는 경우 price-1이 없음. price-2만 있으니 그거 가져오기.
            price2 = driver.find_element(By.CLASS_NAME, 'price-2').text.replace('\n원', "")
            price1_list.append(price2)

        # 할인 후 가격 가져오기
        try:
            price2 = driver.find_element(By.CLASS_NAME, 'price-2').text.replace('\n원', "")
            price2_list.append(price2)
        except:
            price2_list.append(np.nan)

        # 평균평점 가져오기
        evg_review = driver.find_element(By.XPATH, '//*[@id="repReview"]/b').text
        evg_review_list.append(evg_review)

        # 리뷰 클릭하기
        try:
            review_class = driver.find_element(By.CLASS_NAME, 'goods_reputation')
            review_class.click()
            time.sleep(5)
        except:
            pass

        # 리뷰수 가져오기
        try:
            cnt_review = driver.find_element(By.XPATH, '//*[@id="reviewInfo"]/a/span').text.replace('(', "").replace(')', "")
            cnt_review_list.append(cnt_review)
        except:
            cnt_review_list.append(np.nan)

        # 리뷰점수비율 가져오기
        try:
            review_scores = driver.find_elements(By.XPATH, '//*[@id="gdasContentsArea"]/div/div/div/div[3]/ul/li/span[1]')
            review_score_5_list.append(review_scores[0].text)
            review_score_4_list.append(review_scores[1].text)
            review_score_3_list.append(review_scores[2].text)
            review_score_2_list.append(review_scores[3].text)
            review_score_1_list.append(review_scores[4].text)
        except:
            review_score_5_list.append(np.nan)
            review_score_4_list.append(np.nan)
            review_score_3_list.append(np.nan)
            review_score_2_list.append(np.nan)
            review_score_1_list.append(np.nan)

        # 설문조사유형 가져오기
        try:
            poll_type = driver.find_elements(By.XPATH, '//*[@id="gdasContentsArea"]/div/div/dl/dd/ul/li/em')
            poll_type_list1_1.append(poll_type[0].text)
            poll_type_list1_2.append(poll_type[1].text)
            poll_type_list1_3.append(poll_type[2].text)
            poll_type_list2_1.append(poll_type[3].text)
            poll_type_list2_2.append(poll_type[4].text)
            poll_type_list2_3.append(poll_type[5].text)
            poll_type_list3_1.append(poll_type[6].text)
            poll_type_list3_2.append(poll_type[7].text)
            poll_type_list3_3.append(poll_type[8].text)
        except:
            poll_type_list1_1.append(np.nan)
            poll_type_list1_2.append(np.nan)
            poll_type_list1_3.append(np.nan)
            poll_type_list2_1.append(np.nan)
            poll_type_list2_2.append(np.nan)
            poll_type_list2_3.append(np.nan)
            poll_type_list3_1.append(np.nan)
            poll_type_list3_2.append(np.nan)
            poll_type_list3_3.append(np.nan)

    except Exception as e:
        print(f'Error: {e}')

# all_url 파일에서 url 불러온 후 함수 실행
for index, row in all_url.iterrows():  # URL 리스트를 가져옴
    product_id = row['제품id'] # all_url 파일에서 제품id 불러오기
    product_url = row['url'] # all_url 파일에서 url 불러오기
    product_id_list.append(product_id) # 제품id를 product_id_list에 추가
    get_product_info(product_url) # 함수 반복 실행 - 반복 실행하면서 함수 외부의 리스트에 정보가 추가됨

```

```python
print(
      len(product_id_list),
      len(category_list1),
      len(category_list2),
      len(category_list3),
      len(prd_name_list),
      len(price1_list),
      len(price2_list),
      len(evg_review_list),
      len(cnt_review_list),
      len(review_score_5_list),
      len(review_score_4_list),
      len(review_score_3_list),
      len(review_score_2_list),
      len(review_score_1_list),
      len(poll_type_list1_1),
      len(poll_type_list1_2),
      len(poll_type_list1_3),
      len(poll_type_list2_1),
      len(poll_type_list2_2),
      len(poll_type_list2_3),
      len(poll_type_list3_1),
      len(poll_type_list3_2),
      len(poll_type_list3_3)
)
```

```python
# 데이터프레임으로 변환
product_info_df = pd.DataFrame({
    '제품id' : product_id_list,
    'category1': category_list1,
    'category2': category_list2,
    'category3': category_list3,
    'prd_name': prd_name_list,
    'price1': price1_list,
    'price2': price2_list,
    'evg_review': evg_review_list,
    'cnt_review': cnt_review_list,
    'review_score_5': review_score_5_list,
    'review_score_4': review_score_4_list,
    'review_score_3': review_score_3_list,
    'review_score_2': review_score_2_list,
    'review_score_1': review_score_1_list,
    '건성에좋아요': poll_type_list1_1,
    '복합성에좋아요': poll_type_list1_2,
    '지성에좋아요': poll_type_list1_3,
    '보습에좋아요': poll_type_list2_1,
    '진정에좋아요': poll_type_list2_2,
    '주름/미백에좋아요': poll_type_list2_3,
    '자극에순해요': poll_type_list3_1,
    '보통이에요': poll_type_list3_2,
    '자극이느껴져요': poll_type_list3_3
})


# csv로 저장
product_info_df.to_csv('C:/Users/82108/Desktop/final/product_info2.csv')
```
