# 감성분석 함수

```python
from konlpy.tag import Okt
# 형태소 분석기
okt = Okt()

# SentiWord_Dict.txt 감성사전 불러오기
def load_senti_dict(file_path):
    senti_dict = {}
    with open(file_path, 'r', encoding='utf-8') as f:
        for line in f:
            parts = line.strip().split()  # 공백으로 분리    
            if len(parts) > 1:
                word = ' '.join(parts[:-1])  # 마지막 항목을 제외한 나머지를 단어로 결합
                score = int(parts[-1])  # 마지막 항목을 점수로 처리
                senti_dict[word] = score
            else:
                print(f"해당 줄을 무시합니다: {line}")
    return senti_dict

# 감성 사전 로드
senti_dict = load_senti_dict(r'C:\Users\82108\Downloads\KnuSentiLex-master(1)\KnuSentiLex-master\SentiWord_Dict.txt')

# 감성 분석 함수 
def analyze_sentiment(text, senti_dict):
    # 형태소 분석
    morphs = okt.morphs(text)
    
    # 감정 점수 계산
    sentiment_score = 0
    for word in morphs:
        if word in senti_dict:  # 단어가 사전에 있는 경우
            sentiment_score += senti_dict[word]
    
    # 점수에 따라 감정 판단
    if sentiment_score > 0:
        return "긍정"
    elif sentiment_score < 0:
        return "부정"
    else:
        return "중립"
```

## 감성분석_crc
- suncare1 df
- 감성분석 사전 함수
- df -> list
- list를 함수에 적용
- 각 문장마다 긍정/부정/중립 
- 그 결과 list를 df로 저장 

```python
# suncare 1~9 # suncare1_all_crc_df

text_lists = {}

# suncare 1부터 9까지의 파일들을 처리합니다.
for i in range(1, 10):


    # df를 list로 변경하여 딕셔너리에 저장
    text_list = globals()[f'suncare{i}_all_crc_df'].iloc[:, 0].astype(str).tolist()
    text_lists[f'suncare{i}_list'] = text_list

    # 감정 분석 수행
    results = [(text, analyze_sentiment(text, senti_dict)) for text in text_list]

    # 결과 데이터프레임 생성
    results_df = pd.DataFrame(results, columns=['Text', 'Sentiment'])
    globals()[f'suncare{i}_crc_results_df'] = results_df
    print(f'Results for suncare1:')
    print(suncare1_crc_results_df.head())  # 각 결과의 일부를 출력
```

## 긍정/부정/중립 리스트
- 긍정/부정/중립 빈 딕셔너리 생성
- suncare1_list에 각 문장을 감성분석 수행 후 딕셔너리에 저장 
    - 긍정딕셔너리 {'suncare1': 문장들487개}
    - 부정딕셔너리 {'suncare1': 문장들65개}
    - 중립딕셔너리 {'suncare1': 문장들1835개}
- 딕셔너리에 key별로 dataframe으로 변환
- suncare1_crc_positive_df...9
- suncare1_crc_negative_df...9
- suncare1_crc_neutral_df...9

```python
# 딕셔너리 초기화
suncare_crc_positive_dic = {}
suncare_crc_negative_dic = {}
suncare_crc_neutral_dic = {}

# suncare 1부터 9까지 각 리스트에 대해 감정 분석 수행
for i in range(1, 10):
    # 리스트 불러오기
    current_list = text_lists[f'suncare{i}_list']
    
    # 각 리스트에 대한 감정 분석
    positive_list = []
    negative_list = []
    neutral_list = []
    for text in current_list:
        sentiment = analyze_sentiment(text, senti_dict)
        if sentiment == "긍정":
            positive_list.append(text)
        elif sentiment == "부정":
            negative_list.append(text)
        else:
            neutral_list.append(text)
    
    # 딕셔너리에 결과 저장
    suncare_crc_positive_dic[f'suncare{i}'] = positive_list
    suncare_crc_negative_dic[f'suncare{i}'] = negative_list
    suncare_crc_neutral_dic[f'suncare{i}'] = neutral_list

# 결과 확인 (예시로 긍정 딕셔너리에서 suncare1의 결과 출력)
print('긍정:', suncare_crc_positive_dic['suncare1'])
print('부정:', suncare_crc_negative_dic['suncare1'])
print('중립:', suncare_crc_neutral_dic['suncare1'])
```

### 데이터프레임으로 생성 후 excel파일로 저장

```python
for i in range(1, 10):
    # 결과 데이터프레임 생성  
    suncare_crc_positive_df = pd.DataFrame(suncare_crc_positive_dic[f'suncare{i}'], columns=['text'])
    suncare_crc_negative_df = pd.DataFrame(suncare_crc_negative_dic[f'suncare{i}'], columns=['text'])
    suncare_crc_neutral_df = pd.DataFrame(suncare_crc_neutral_dic[f'suncare{i}'], columns=['text'])
    
    # 동적으로 DataFrame 할당
    globals()[f'suncare{i}_crc_positive_df'] = suncare_crc_positive_df
    globals()[f'suncare{i}_crc_negative_df'] = suncare_crc_negative_df
    globals()[f'suncare{i}_crc_neutral_df'] = suncare_crc_neutral_df
    
    # Excel 파일로 저장
    suncare_crc_positive_df.to_excel(f'suncare{i}_crc_positive.xlsx', index=False)
    suncare_crc_negative_df.to_excel(f'suncare{i}_crc_negative.xlsx', index=False)
    suncare_crc_neutral_df.to_excel(f'suncare{i}_crc_neutral.xlsx', index=False)
```
